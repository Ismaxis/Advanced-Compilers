```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
i2 = 0
i4 = 1
i5 = 2
i7 = 3

j2 = 4
j3 = 5
j4 = 6
j6 = 7

n2 = 8
n3 = 9

k = 10

labels = {
    0: "i2",
    1: "i4",
    2: "i5",
    3: "i7",
    4: "j2",
    5: "j3",
    6: "j4",
    7: "j6",
    8: "n2",
    9: "n3",
    10: "k",
}

edges = {
    i2: [j2, n2, k, j4, n3],
    i4: [n2, j4, n3, j3],
    i5: [k, j4, j6, i7, n2],
    i7: [i5, k, j4, j6, n2],
    j2: [i2, k, j4, n3, n2],
    j3: [i4, k, j4, n2, n3],
    j4: [n2, n3, j6, j3, i7, i4, i5, i2, k, j2],
    j6: [i7, i5, k, j4, n2],
    n2: [i2, i4, j2, k, j4, j3, j6, i7, i5],
    n3: [j2, i2, j4, i4, j3],
    k: [j2, n2, j4, j6, j3, i7, i5, i2],
}

edges2 = set()

for v, es in edges.items():
    for u in es:
        assert edges[u].count(v) > 0, f"{u}, {v}"
        assert u != v

        edges2.add((v, u))
        edges2.add((u, v))


G = nx.DiGraph()

G.add_edges_from(edges2)

color_map = nx.coloring.greedy_color(G)
print(max(color_map.values())+1)

# pos = nx.spring_layout(G)
# nx.draw(G, pos, node_color=[color_map[c] for c in G.nodes()], labels=labels)
# plt.show()
```
